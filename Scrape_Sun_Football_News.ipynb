```python
#ChatGPT for the win finally 
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

website = "https://www.thesun.co.uk/sport/football/"
path = "C:/Users/Dell/Python Automation/chromedriver-win64/chromedriver.exe"
service = Service(executable_path=path)
driver = webdriver.Chrome(service=service)
driver.get(website)

# Wait for page to fully load and add a small delay
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.XPATH, "//div[@class='teaser__copy-container']")))
time.sleep(2)  # Add a short delay to ensure dynamic content loads

# Get all article containers
containers = driver.find_elements(By.XPATH, "//div[@class='teaser__copy-container']")

titles = []
subtitles = []
links = []

for container in containers:
    try:
        # Extract title from span with class 'teaser__headline'
        title_element = container.find_element(By.XPATH, ".//span[@class='teaser__headline teaser__kicker t-p-color']")
        title = title_element.get_attribute("data-original-text") or title_element.text
        print(f"Extracted Title: '{title}'")
    except Exception as e:
        title = ""
        print(f"Title not found: {str(e)[:100]}...")
    
    try:
        # Extract subtitle from h3 with class 'teaser__subdeck'
        subtitle_element = container.find_element(By.XPATH, ".//h3[@class='teaser__subdeck']")
        subtitle = subtitle_element.get_attribute("data-original-text") or subtitle_element.text
        print(f"Extracted Subtitle: '{subtitle}'")
    except Exception as e:
        subtitle = ""
        print(f"Subtitle not found: {str(e)[:100]}...")
    
    try:
        # Get the link from the a tag
        link_element = container.find_element(By.XPATH, ".//a")
        link = link_element.get_attribute("href")
        print(f"Extracted Link: '{link}'")
    except Exception as e:
        link = ""
        print(f"Link not found: {str(e)[:100]}...")
    
    titles.append(title)
    subtitles.append(subtitle)
    links.append(link)

driver.quit()

# Create DataFrame and display results
df = pd.DataFrame({'title': titles, 'subtitle': subtitles, 'link': links})
print(df)

df.to_csv('thesun_football_articles.csv', index=False)

# Save to CSV (optional)

```
